```python
# Import Libraries
!pip install mne
import mne
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load the signal
data = mne.io.read_raw_edf("h01.edf", preload=True)
```

```python
# Extract information from the signal
n_channels = len(data.ch_names)
S_freq = data.info['sfreq']
cha_names = data.ch_names
data_shape = data.get_data().shape
# Print the extracted information
print("Number of channels:", n_channels)
print("Sampling Frequency:", S_freq)
print("Channel Names:", cha_names)
print("Data Shape:", data_shape)
```

```python
# Signal Visualization
data.plot()
plt.show()
```

```python
# Plotting first 5 second of the signal
data.plot(duration = 5, n_channels=19)
plt.show()
```
